```python
import pandas as pd  
import os
import emoji
```

```python
def open_files(root_dir):
    csv_file_paths = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".csv"):
                file_path = os.path.join(root,file)
                csv_file_paths.append(file_path)
    return csv_file_paths
```

```python
len(open_files(root_directory))
```

```python
merged_csv = pd.DataFrame()
for file in open_files(root_directory):
    df = pd.read_csv(file)
    merged_csv = pd.concat([merged_csv,df])
```

```python
merged_csv.info()
```

```python
merged_csv.to_csv('merged_file.csv', index=False)
```

```python
emoji_count = {}

for comment in merged_csv["Comment Text"].astype(str):
    for char in comment:
        if emoji.is_emoji(char):
            if char in emoji_count:
                emoji_count[char]+=1
            else:
                emoji_count[char]=1
print(emoji_count)
```

```python
emoji_count=dict(sorted(emoji_count.items(), key=lambda item: item[1],reverse=True))
emoji_count
```

```python
comment_2015=pd.DataFrame(merged_csv[merged_csv['Year']=='2015']['Comment Text'])
comment_2015['Comment Text'] = comment_2015['Comment Text'].astype(str)
```

```python
from langdetect import LangDetectException, detect

language_counts = {}

for sentence in comment_2015['Comment Text']:
    if isinstance(sentence, str) and len(sentence.strip()) > 2:
        try:
            language = detect(sentence)
            if language in language_counts:
                language_counts[language] += 1
            else:
                language_counts[language] = 1
        except LangDetectException:
            pass  

sorted_language_counts = dict(sorted(language_counts.items(), key=lambda item: item[1], reverse=True))

for language, count in sorted_language_counts.items():
    print(language,count)
```

# Abbreviations

```python
path = "C:\\Users\\sukan\\OneDrive\\Desktop\\Python\\FBcomments\\Taylor_Swift\\Abbreviations.txt"

with open(path,'r') as file: 
    abb = file.read()
```

```python
abbreviations = []
for l in abb.split('\n'):
    for w in l.split(':'):
        if len(w)<= 7:
            abbreviations.append(w.lower())
len(abbreviations)
```

```python
merged_csv['Comment Text'] = merged_csv['Comment Text'].astype(str)
```

```python
merged_csv['Comment Text']=merged_csv['Comment Text'].apply(lambda x: x.lower())   
```

```python
abb_found = []
for comment in merged_csv['Comment Text']:
    for w in comment.split():
        for i in abbreviations:
            if w==i:
                abb_found.append(i)
len(abb_found)
```

```python
len(set(abb_found))
```

```python
abb_counter ={}

for term in abb_found:
    if term in abb_counter:
        abb_counter[term]+=1
    else:
        abb_counter[term]=1
```

```python
abb_counter = dict(sorted(abb_counter.items(), key=lambda item: item[1], reverse=True))
```
