# year 2018 in 2018post

```python
c18_p18=pd.DataFrame(df[df['Year']=='2018'][['Year','Comment Text']])
```

```python
c18_p18.info()
```

```python
c18_p18['Comment Text']=c18_p18['Comment Text'].astype(str)
```

```python
sen18 = [i for i in c18_p18['Comment Text']]
sen18 = ' '.join(sen18)

char = [i for i in sen18]
spaced_char = ' '.join(char)
```

```python
#Regex is not helping

import re

pattern = re.findall(r'[^\w\s\d!"$%&\^./|#+()\[\]{}^~—_`?@<=>:;*+,-]', spaced_char)

non_alpha=[i for i in pattern]
len(non_alpha)
```

```python
print(pattern)
```

```python
import emoji

emojis_found = []

for i in char:
    if emoji.is_emoji(i):
        emojis_found.append(i)

print(len(emojis_found))
```

```python
first_emojis = []

for comment in c18_p18['Comment Text']:
    emoji_found = False
    for char in comment:
        if emoji.is_emoji(char):
            first_emojis.append(char)
            emoji_found = True
            break 
    if not emoji_found:
        first_emojis.append(None) 

emojipercommnet = []

for i in first_emojis:
    if emoji.is_emoji(i):
        emojipercommnet.append(i)

print(len(emojipercommnet))
```
